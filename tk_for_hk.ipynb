```python

from time import sleep
from selenium import webdriver
from selenium.webdriver.support.ui import Select

'''
关于selenium与谷歌浏览器的配置参考： https://blog.csdn.net/muriyue6/article/details/101440353

'''


#ddddocr库不需要自己训练模型，准确率也有80%左右。
import ddddocr
import time

#不重要的库
import winsound
import smtplib
from email.mime.text import MIMEText
from email.header import Header
```

```python
#获取今天的日期
a=list(time.localtime())[2]
```

```python
num_url = 'https://hk.sz.gov.cn:8118/userPage/login'

#填入chromedriver.exe路径
path='C:\Program Files\Google\Chrome\Application\chromedriver.exe'

#选择登录的帐号类型
#港澳居民來往內地通行證为 4 ; 臺灣居民來往大陸通行證为 6 ; 往來港澳通行證为 2 ; 护照为 3
type_account=4

#帐号
account="H04657315"

#密码
code="a2034299897"

```

```python
driver = webdriver.Chrome(executable_path=path)
driver.implicitly_wait(30)
driver.get(num_url)
driver.maximize_window()
sleep(1)
url1 = driver.current_url
sleep(1)
driver.find_element_by_xpath('//button[@type="button" and @onclick="closeLoginHint()"]').click()
```

```python
##选择型的题目
selector = Select(driver.find_element_by_id("select_certificate"))
selector.select_by_value("{}".format(type_account))
#sleep(1)

##填入帐号密码
id=driver.find_element_by_id("input_idCardNo")
id.send_keys(account)
#sleep(1)

password=driver.find_element_by_id("input_pwd")
password.send_keys(code)
#sleep(1)
```

```python
#返回对应验证码
def fun():
    img_ = driver.find_element_by_id("img_verify")
    data = img_.screenshot_as_png    #截图的方法是最好的！
    with open( 'code.png','wb' ) as f:
        f.write(data)

    #卷王ocr
    ocr = ddddocr.DdddOcr(old=True)
    with open("code.png", 'rb') as f:
        image = f.read()

    res = ocr.classification(image)
    #print(res)
    return res
```

```python
#用url判断是否跳转了新标签页 
while driver.current_url == url1: 
    
    #无法显示图片，则点击切换
    while fun() == "d":
        driver.find_element_by_id("img_verify").click()
        #sleep(10)
    
    #将验证码填入，错了就切换验证码且继续
    input_box = driver.find_element_by_id("input_verifyCode")
    input_box.clear()
    input_box.send_keys(fun())
    driver.find_element_by_id('btn_login').click()
    sleep(2)
    if driver.current_url == url1:
        sleep(4)
        driver.find_element_by_id("img_verify").click()
    
```

```python
sleep(1)
driver.find_element_by_xpath('//button[@type="button" and @onclick="closeHint()"]').click()
canbook = driver.find_element_by_id("a_canBookHotel")
canbook.click()
sleep(1)
url2 = driver.current_url
```

```python
def fun2():
    img_ = driver.find_element_by_id("img_verify")
    data = img_.screenshot_as_png    
    with open( 'code2.png','wb' ) as f:
        f.write(data)

    ocr = ddddocr.DdddOcr(old=True)
    with open("code2.png", 'rb') as f:
        image = f.read()
    res = ocr.classification(image)
    return res

def check_insert_click():
    while driver.current_url!=url2: 

        while fun2() == "d":
            driver.find_element_by_id("img_verify").click()
            sleep(2)

        #将验证码填入，错了就切换验证码且继续
        input_box = driver.find_element_by_id("checkCode")
        input_box.clear()
        input_box.send_keys(fun2())
        sleep(0.1)
        driver.find_element_by_id("btnSubmit").click()
        sleep(0.5)
        driver.find_element_by_xpath("//div[@class='layui-m-layerbtn']/span[2]").click()
        sleep(2)
        if driver.current_url!=url2:
            sleep(2)
            driver.find_element_by_id("img_verify").click()
    
```

```python
#最核心的万能方法，xpath。从元素的区分点入手，如3号、4号，再找其父节点同级的p节点，p节点子节点即button元素
#check_insert_click()
```

```python

while 1:
    try:
        ff=driver.find_element_by_xpath("//em[contains(text(),'{}')]/../following-sibling::p/button".format(a))
        gg=driver.find_element_by_xpath("//em[contains(text(),'{}')]/../following-sibling::p/button".format(a+1))
        dd=driver.find_element_by_xpath("//em[contains(text(),'{}')]/../following-sibling::p/button".format(a+2))
        ee=driver.find_element_by_xpath("//em[contains(text(),'{}')]/../following-sibling::p/button".format(a+3))
        aa=driver.find_element_by_xpath("//em[contains(text(),'{}')]/../following-sibling::p/button".format(a+4))
        bb=driver.find_element_by_xpath("//em[contains(text(),'{}')]/../following-sibling::p/button".format(a+5))
        cc=driver.find_element_by_xpath("//em[contains(text(),'{}')]/../following-sibling::p/button".format(a+6))

        while (aa.get_attribute("class")=="orange"):
            aa.click()
            sleep(0.2)
            driver.find_element_by_xpath("//button[contains(text(),'確定')]").click()
            check_insert_click()
            break

        while (bb.get_attribute("class")=="orange"):
            bb.click()
            sleep(0.2)
            driver.find_element_by_xpath("//button[contains(text(),'確定')]").click()
            check_insert_click()
            break
        while (cc.get_attribute("class")=="orange"):
            cc.click()
            sleep(0.2)
            driver.find_element_by_xpath("//button[contains(text(),'確定')]").click()
            check_insert_click()
            break
        while (dd.get_attribute("class")=="orange"):
            dd.click()
            sleep(0.2)
            driver.find_element_by_xpath("//button[contains(text(),'確定')]").click()
            check_insert_click()
            break    
        while (ff.get_attribute("class")=="orange"):
            ff.click()
            sleep(0.2)
            driver.find_element_by_xpath("//button[contains(text(),'確定')]").click()
            check_insert_click()
            break
        while (gg.get_attribute("class")=="orange"):
            gg.click()
            sleep(0.2)
            driver.find_element_by_xpath("//button[contains(text(),'確定')]").click()
            check_insert_click()
            break
        while (ee.get_attribute("class")=="orange"):
            ee.click()
            sleep(0.2)
            driver.find_element_by_xpath("//button[contains(text(),'確定')]").click()
            check_insert_click()
            break

        driver.refresh()
        sleep(1)#刷新的间隔    
    except:
        driver.refresh()
```
